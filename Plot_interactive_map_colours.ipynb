```python
import geopandas as gpd
import pandas as pd
import numpy as np

%matplotlib inline


#zones = gpd.read_file('/home/swb1g09/Distribution/Areas.shp').to_crs({'init' :'epsg:4326'})
zones = gpd.read_file('/home/swb1g09/Distribution/Areas_simple.json').to_crs({'init' :'epsg:4326'})

#zones['zone'] = zones.zone.astype('category')
#zones['borough'] = zones.borough.astype('category')
zones.sort_values(by='name',inplace=True)
zones=zones.reset_index(drop=True)
#print(zones.head(n=20))

Countries = pd.read_csv('/home/swb1g09/Distribution/Area_conversion.csv')

#Countries

Countries.head()
zones=zones.merge(Countries,on='name')
zones.head(n=5)
```

```python
import random

counts = pd.DataFrame(zones.name)

counts = counts.sample(10)
counts['count'] = np.random.uniform(0,1,len(counts))

#print(counts.head())

zones=zones.join(counts.set_index('name'),on='name')

zones
```

```python
from bokeh.io import output_notebook
from bokeh.models import GeoJSONDataSource, HoverTool, CategoricalColorMapper, LinearColorMapper
from bokeh.plotting import figure, show
from bokeh.palettes import Category10
from bokeh.palettes import viridis
from bokeh.palettes import RdYlGn
from bokeh.models import ColorBar, LogColorMapper

output_notebook()
```

```python
zonesWOdata = zones[zones.isnull().any(axis=1)]
zonesWdata = zones[np.invert(zones.isnull().any(axis=1))]
```

```python

geo_sourceWOdata = GeoJSONDataSource(geojson=zonesWOdata.to_json())
geo_sourceWdata = GeoJSONDataSource(geojson=zonesWdata.to_json())

#factors = zones.Country.drop_duplicates()
#print(factors)

color_mapper = LogColorMapper(RdYlGn[11])

fig = figure(match_aspect=True)
fig.patches(xs='xs', ys='ys', alpha=1., source=geo_sourceWdata,
            color={'field': 'count', 'transform': color_mapper}, 
            line_width=1, line_color='black')

fig.patches(xs='xs', ys='ys', alpha=1., source=geo_sourceWOdata,
            fill_color="white", 
            line_width=1, line_color='black')

#color_mapper = CategoricalColorMapper(factors=factors.tolist(), palette=Category10[10])
#fig.patches(xs='xs', ys='ys', alpha=0.9, source=geo_source, line_color='#000000',
#          color={'field': 'Country', 'transform': color_mapper},
#         )

hover = HoverTool(
    point_policy='follow_mouse',
    tooltips='<p><b>Country</b>: @Country</p><p><b>Name</b>: @long_name</p>'
)
fig.add_tools(hover)


color_bar = ColorBar(color_mapper=color_mapper, 
                     location=(0, 0), 
                     label_standoff=32)
fig.add_layout(color_bar, 'right')

fig.xaxis.visible = False
fig.yaxis.visible = False
fig.grid.visible = False

show(fig)

```
